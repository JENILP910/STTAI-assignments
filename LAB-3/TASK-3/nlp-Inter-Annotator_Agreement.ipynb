# 1.Cohen‚Äôs Kappa for the NLP Dataset Task

```python
def create_sentence_labels(data):
    sentences_data = {}

    for entry in data:
        sentence_text = entry["data"]["text"]
        # print(sentence_text)
        annotations = entry["annotations"][0]["result"]  # Assuming 1 annotation per sentence
        # print(annotations)
        
        # Extract tokens for annotator 1
        word_labels = {
            # {"text": ann["value"]["text"], "label": ann["value"]["labels"][0]}
            ann["value"]["text"] : ann["value"]["labels"][0]
            for ann in annotations
        }

        sentences_data[sentence_text] = word_labels
    return sentences_data
```

```python
import json

# Load the data
with open(datasetDir.__str__()+"\JENIL_POS.json", "r", encoding="utf-8") as f:
    data = json.load(f)

dataset1 = create_sentence_labels(data)

# Load the second data
with open(datasetDir.__str__()+"\Mithlesh_POS.json", "r", encoding="utf-8") as f:
    data = json.load(f)

dataset2 = create_sentence_labels(data)
```

```python
# printing data
for sentence, word_labels in dataset1.items():
    print(sentence)
    print(word_labels)
    break
```

```python
# For observed_agreement

for (sentence_1, word_labels_1), (sentence_2, word_labels_2) in zip(dataset1.items(), dataset2.items()):
    print(f"Word 1: {sentence_1}, Labels 1: {word_labels_1}")
    print(f"Word 2: {sentence_2}, Labels 2: {word_labels_2}")
    break
```

```python
# For observed agreement
observed_agreement = {}

for (sentence_1, word_labels_1), (sentence_2, word_labels_2) in zip(dataset1.items(), dataset2.items()):
    # print(f"Word 1: {sentence_1}, Labels 1: {word_labels_1}")
    # print(f"Word 2: {sentence_2}, Labels 2: {word_labels_2}")
    if (sentence_1 == sentence_2):
        # Clean the words by stripping any leading/trailing spaces before getting the keys
        word_labels_1 = {k.strip(): v for k, v in word_labels_1.items()}
        word_labels_2 = {k.strip(): v for k, v in word_labels_2.items()}

        # Inner join: Get common words between annotators and compare labels
        common_word_labels = set(word_labels_1.keys()) & (set(word_labels_2.keys()))

        joined_data = {}
        for word in common_word_labels:
            labels_1 = word_labels_1[word]
            labels_2 = word_labels_2[word]
            
            # Checking if the labels are the same
            joined_data[word] = labels_1 == labels_2

        # print(joined_data)
        observed_agreement[sentence_1] = joined_data
    # break
```

```python
# visualizing output
from tabulate import tabulate

for sentence, word_choice in observed_agreement.items():
    print(f"Sentence: {sentence}")

    # Printing the table
    print(tabulate(list(word_choice.items()), headers=["Words", "Is Same"], tablefmt="grid", colalign=("left", "center")), "\n")
        
```

```python
from collections import Counter


# Calculate Observed Agreement (P0)
# Initialize counters
true_count = 0
false_count = 0
total_count = 0

# Iterate through the dictionary
for key, value_dict in observed_agreement.items():
    for val in value_dict.values():
        if val:
            true_count += 1
        else:
            false_count += 1
        total_count += 1


print(f"True count: {true_count}")
print(f"False count: {false_count}")

P0 = true_count / total_count




# Calculate Expected Agreement (Pe)
# Flatten the word-labels from all sentences
labels_1 = [label for sentence in dataset1.values() for word, label in sentence.items()]
labels_2 = [label for sentence in dataset2.values() for word, label in sentence.items()]

# Count the occurrences of each label for both annotators
count_1 = Counter(labels_1)
count_2 = Counter(labels_2)
print(f"\nAnnotator 1 labels: {count_1}")
print(f"Annotator 2 labels: {count_2}")

# Total number of labels
total_labels_1 = len(labels_1)
total_labels_2 = len(labels_2)
print(f"\nAnnotator 1 total labels marked: {total_labels_1}")
print(f"Annotator 2 total labels marked: {total_labels_2}")

# Calculate Pe using the formula for expected agreement
Pe = 0
for label in count_1:
    prob_1 = count_1[label] / total_labels_1
    prob_2 = count_2.get(label, 0) / total_labels_2
    Pe += prob_1 * prob_2


# Step 3: Calculate Cohen's Kappa
kappa = (P0 - Pe) / (1 - Pe)

# Output results
print(f"\nObserved Agreement (P0): {P0:.4f}")
print(f"Expected Agreement (Pe): {Pe:.4f}")
print(f"\nCohen's Kappa: {kappa:.4f}")
```

```python
import pandas as pd

cv_mithlesh=pd.read_csv("Mithlesh_CV.csv")
cv_jenil=pd.read_csv("JENIL_CV.csv")
cv_outsider=pd.read_csv("py.csv")
```

```python
from statsmodels.stats.inter_rater import fleiss_kappa



# Extract the two columns
choice_column = cv_mithlesh["choice"]
second_column = cv_jenil["choice"]

# Organize data into a rating matrix
# Assuming categories are integers or labels, create the rating matrix
categories = sorted(set(choice_column) | set(second_column))  # All unique categories
rating_matrix = []

for i in range(len(choice_column)):
    row = [0] * len(categories)
    row[categories.index(choice_column[i])] += 1
    row[categories.index(second_column[i])] += 1
    rating_matrix.append(row)

# Convert to DataFrame for clarity (optional)
rating_df = pd.DataFrame(rating_matrix, columns=categories)

# Calculate Fleiss' Kappa
kappa = fleiss_kappa(rating_df.values, method='fleiss')
print("Fleiss' Kappa:", kappa)

rating_df
```
